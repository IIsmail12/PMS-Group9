```python
import matplotlib.pyplot as plt

# --- Graph 1: Market Share 2025 ---
labels = ["T-Mobile", "Verizon", "AT&T", "Others (MVNO)"]
market_share = [35, 34, 27, 4]
colors = ["pink", "red", "royalblue", "gray"]

plt.figure(figsize=(7,7))
plt.pie(market_share, labels=labels, autopct="%1.1f%%", startangle=90, colors=colors)
plt.title("U.S. Wireless Market Share (2025)  Big Three", fontsize=14)
plt.show()

# --- Graph 2: Estimated Subscribers 20202024 ---
years = [2020, 2021, 2022, 2023, 2024]
verizon_subs = [130, 136, 140, 144, 146]  # estimate ending 146M
att_subs     = [110, 114, 117, 119, 120]  # estimate ending 120M

plt.figure(figsize=(10,5))
plt.plot(years, verizon_subs, marker="o", color="red", linewidth=2, label="Verizon")
plt.plot(years, att_subs,     marker="o", color="royalblue", linewidth=2, label="AT&T")
plt.title("Wireless Subscribers (Estimate): Verizon vs AT&T (20202024)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Subscribers (Millions)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Companies
companies = ["AT&T", "Verizon"]

# Customer segmentation (millions)
consumer = np.array([85, 100])
business = np.array([20, 28])

x = np.arange(len(companies))
width = 0.35  # width of each bar

plt.figure(figsize=(10,6))

# Bars for Consumer
plt.bar(x - width/2, consumer, width, label="Consumer Wireless", color="navy")

# Bars for Business
plt.bar(x + width/2, business, width, label="Business Wireless", color="grey")

plt.xticks(x, companies)
plt.ylabel("Customers (Millions)")
plt.title("Wireless Customer Segmentation (2024): Consumer vs Business")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()

```
